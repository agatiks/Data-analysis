## Final features

```python
def minmax(data, column):
    data[column] = (data[column] - np.min(data[column]))/(np.max(data[column]) - np.min(data[column]))
def target_scaling(data):
    data[target] = np.log(data[target])
def target_unscaling(data):
    np.exp(data)
```

```python
def make_features(data,  test = False, dop = None):
    new_data = data.copy()
    target_scaling(new_data)
    features = ['osm_city_nearest_population']
    if not test:
        new_data.drop(new_data[new_data[target] > 250000].index, inplace = True)
    minmax(new_data, features[0])
    new_data[features[0]].fillna(new_data[features[0]].mean(), inplace = True) #    nan
    return new_data, features
def make_selection(data, test = False, gr = False, dop = None):
    new_data, features = make_features(data, test = test, dop = dop)
    if not test:
        if gr:
            features_y = features + [target]
            sns.heatmap(new_data[features_y].corr().abs(), vmin=0, vmax=0.6, cmap="BuPu")
            print(new_data[features_y].corr()[target])
        return new_data[features], new_data[target]
    else:
        return new_data[features]
```

## Learning

```python
train_X, train_y = make_selection(train, gr = True)
```

```python
train_X.describe()
```

```python
val_X, val_y = make_selection(val, dop = train)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_absolute_percentage_error
```

```python
model = SGDRegressor()
```

```python
model.fit(train_X, train_y)
pred = model.predict(val_X)
target_unscaling(pred)
```

```python
mape = mean_absolute_percentage_error(pred, val_y)
print(mape)
```
